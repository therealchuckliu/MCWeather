```python
import pandas as pd
import requests
from datetime import date
import time
import os.path
```

```python
def fetch_csv(region, startdate, enddate):
    frames = []
    for year in range(startdate.year, enddate.year + 1):
        try:
            frame = pd.read_csv('data/{}-{}.csv'.format(region, year))
            if startdate.year == year:
                startidx = frame[frame.EST=="{}-{}-{}".format(startdate.year, startdate.month, startdate.day)].index[0]
                frame = frame[frame.index >= startidx]
            if enddate.year == year:
                endidx = frame[frame.EST=="{}-{}-{}".format(enddate.year, enddate.month, enddate.day)].index[0]
                frame = frame[frame.index <= endidx]
            frames.append(frame)
        except:
            return False
    return pd.concat(frames)

def fetch_data(region, startdate, enddate):
    frame = fetch_csv(region, startdate, enddate)
    if type(frame) is bool:
        # data not in data folder, grab yearly data and save it
        for year in range(startdate.year, enddate.year+1):
            if not os.path.isfile("data/{}-{}.csv".format(region, year)):
                with open('data/{}-{}.csv'.format(region, year), 'w') as wfile:
                    url = "https://www.wunderground.com/history/airport/" + \
                    "{}/{}/{}/{}/CustomHistory.html?".format(region, year, 1, 1) + \
                    "dayend={}&monthend={}&yearend={}&format=1".format(31, 12, year)
                    csv_data = requests.get(url).text.replace("<br />", "").strip(' \t\n\r')
                    wfile.write(csv_data)
                time.sleep(2)
        return fetch_csv(region, startdate, enddate)
    return frame
```

```python
fetch_data("KBOS", date(2010, 5, 15), date(2012, 5, 15))
```
